## *Seteo de parametros de la notebook*

```python
# Semilla para el .sample()
seed = random.randint(10,500)
foco_muestra = "baja_presencia"  # "alta_presencia" , "baja_presencia" o "sin_foco"
if foco_muestra not in ["alta_presencia" , "baja_presencia", "sin_foco"]:
    raise ValueError('Revisar el valor que le pasamos al parametro "foco_muestra", las opciones disponibles son: "alta_presencia", "baja_presencia" o "sin_foco"')
print(f'Parametros de la simulación: seed:{seed}, foco_muestra:{foco_muestra}')
```

## 1) Pre procesing

**Removing duplicates**

```python
# Tratamiento de la feautre "runtime"
df_scoreo_oot['runtime'] = pd.to_datetime(df_scoreo_oot.runtime.apply(lambda x: str(x)[:19]))
df_scoreo_oot = df_scoreo_oot.sort_values(['runtime'])
# Ahora si, teniendo el df ordenado, eliminamos los duplicados, quedandonos con el registro más reciente (la última ocurrencia)
df_scoreo_oot.drop_duplicates(subset=['car_id'], keep='last', inplace=True)
```

```python
df_scoreo_oot.shape
```

**Removing outliers**

```python
# Construcción del precio final
blue= 196.5
oficial= 108.4
col1 = 'price_symbol'
col2 = 'car_kms'
conditions = [df_scoreo_oot[col1]!='U$S', (df_scoreo_oot[col1]=='U$S') & (df_scoreo_oot[col2]==0), (df_scoreo_oot[col1]=='U$S') & (df_scoreo_oot[col2]!=0)]
choices = [df_scoreo_oot.price_amount, df_scoreo_oot['price_amount']*oficial, df_scoreo_oot['price_amount']*blue]

df_scoreo_oot['price_meli_ok'] = np.select(conditions, choices, default=np.nan)
```

```python
# Luego de ya tener el precio final calculado, podemos eliminar del dataset las variables de precio originales
df_scoreo_oot.drop(['price_symbol','price_amount'],1,inplace=True)
```

```python
# Analisis: Excluir outliers de "price_meli_ok"
q_99_p = np.quantile(df_scoreo_oot['price_meli_ok'],.99)
q_99_km = np.quantile(df_scoreo_oot['car_kms'],.99)
df_scoreo_oot = df_scoreo_oot[(df_scoreo_oot.price_meli_ok <= q_99_p) & (df_scoreo_oot.car_kms <= q_99_km)]
```

**Removing catgories**

```python
cats_loc1_ok = ['Capital Federal','Bs.As. G.B.A. Sur','Bs.As. G.B.A. Norte'
           ,'Bs.As. G.B.A. Oeste','Buenos Aires Interior','Bs.As. Costa Atlántica']
mask1 = df_scoreo_oot.car_location_1.apply(lambda x: x in cats_loc1_ok)
cats_brand_ok =['Alfa Romeo','Audi','BMW','Chery','Chevrolet','Citroën','DS','Fiat','Ford','Geely','Haval','Honda',
 'Hyundai','JAC','Jeep','Kia','Land Rover','Lifan','Mercedes-Benz','Mini','Mitsubishi','Nissan','Peugeot','Porsche','RAM',
 'Renault','SEAT','Shineray','Smart','Subaru','Suzuki','Toyota','Volkswagen','Volvo'] # Exclusion x pedido Edgar: Dodge, Chrysler
mask2 = df_scoreo_oot.brand.apply(lambda x: x in cats_brand_ok)
sub_mask = df_scoreo_oot.brand == 'Jeep'
lst_not_renegade = [i for i in df_scoreo_oot[sub_mask]['match_modelo_a'].unique() if i != 'Renegade']
mask3 = df_scoreo_oot.match_modelo_a.apply(lambda x: x not in lst_not_renegade)
df_scoreo_oot = df_scoreo_oot[(mask1 & mask2 & mask3)]
```

**Removing 0km & concesionaria**

```python
# Pasamos "dealer" de booleano a flag 1 y 0
df_scoreo_oot['dealer'] = np.where(df_scoreo_oot['dealer']==True,1,0)
mask_not_0km = df_scoreo_oot.car_kms > 90
mask_not_conces = df_scoreo_oot.dealer == 0

df_scoreo_oot = df_scoreo_oot[(mask_not_0km) & (mask_not_conces)]

# Como nos quedamos con solamente un valor de la feature "dealer" la podemos eliminar
df_scoreo_oot.drop('dealer',1,inplace=True)
```

**Removing match_scores under 50%**

```python
lst = ['score_marca_a','score_modelo_a','score_v1_c']
for col in lst:
    df_scoreo_oot = df_scoreo_oot[df_scoreo_oot[col]>=50]
    
# chequeo
df_scoreo_oot[lst].min()
```

**Selecting final feautres**

```python
final_features = ['car_year','car_kms','match_marca_a','match_modelo_a'
                  ,'match_v1_c', 'Subseg_a', 'Seg_a']
car_location = ['car_location_1', 'match_v1_a'] # estas no las usamos para entrenar el modelo, solo que la queremos en el output final
car_ids = ['car_id'] # lo necesitamos para que las simulaciones no scoreen registros que ya se scorearon antes
target = ['price_meli_ok'] # lo necesitamos para calcular el precio_promedio, el cual va a ser parte del output final
df_scoreo_oot_final = df_scoreo_oot[final_features+car_location+car_ids+target]
```

```python
# esto simplemente lo hacemos para despues en el catboost diferenciar entre float y object, en lugar de sumar int. 
df_scoreo_oot_final['car_year']=df_scoreo_oot_final['car_year'].astype('float')
```

**Removing nulls**

```python
# Eliminamos los registros con nulos
for col in final_features:
    df_scoreo_oot_final = df_scoreo_oot_final[~df_scoreo_oot_final[col].isnull()]
```

```python
df_scoreo_oot_final.shape
```

**Foco de la muestra**

En este step lo que sucede es que filtramos la muestra para quedarnos con un grupo selecto en base a la presencia en MELI. Ejemplo: si seteamos foco_muestra = 'alta_presencia', nuestro dataset va a quedar compuesto únicamente de los modelos con más presencia en la plataforma.

```python
if foco_muestra == 'alta_presencia':
    df_grouped = df_scoreo_oot_final.groupby(['match_marca_a','match_modelo_a'], as_index=False).size()
    df_grouped= df_grouped.sort_values(by='size', ascending=False).iloc[:15,:]  # los primeros 15 registros
    marca_cats = list(df_grouped.match_marca_a.unique())
    modelo_cats = list(df_grouped.match_modelo_a.unique())
    mask1 = df_scoreo_oot_final.match_marca_a.apply(lambda x: x in marca_cats)
    mask2 = df_scoreo_oot_final.match_modelo_a.apply(lambda x: x in modelo_cats)
    df_scoreo_oot_final = df_scoreo_oot_final[(mask1 & mask2)]
elif foco_muestra == 'baja_presencia':
    df_grouped = df_scoreo_oot_final.groupby(['match_marca_a'], as_index=False).size()
    df_grouped= df_grouped.sort_values(by='size', ascending=False).iloc[-15:,:]  # los ultimos 15 registros
    marca_cats = list(df_grouped.match_marca_a.unique())
    mask = df_scoreo_oot_final.match_marca_a.apply(lambda x: x in marca_cats)
    df_scoreo_oot_final = df_scoreo_oot_final[mask]
elif foco_muestra == 'sin_foco':
    pass
else:
    raise ValueError('Revisar el valor que le pasamos al parametro "foco_muestra", las opciones disponibles son "alta_presencia", "baja_presencia" o "sin_foco"')
```

```python
df_scoreo_oot_final.shape
```

## 2) Seleccion muestra final (50 casos)

```python
# Codigo para hacer el check de que los registros que estemos scoreando, no los hayamos scoreado en simulaciones previas
# Por ahora no es necesario, por eso lo dejamos comentado

# Levantamos los ids ya scoreados en anteriores simulaciones, para no volverlos a scorear
#df_ids_already_scored = pd.read_excel('df_ids_already_scored.xlsx',engine='openpyxl', index_col=[0])
#ids_to_exclude = df_ids_already_scored.ids_already_scored.unique().tolist()

#df_scoreo_oot_final.car_id.apply(lambda x: x not in ids_to_exclude).sum()

# tienen que ser 50 menos que la anterior
#mask_ids = df_scoreo_oot_final.car_id.apply(lambda x: x not in ids_to_exclude)
#df_scoreo_oot_final[mask_ids].shape

#mask_ids = df_scoreo_oot_final.car_id.apply(lambda x: x not in ids_to_exclude)
#df_scoreo_oot_final_sample = df_scoreo_oot_final[mask_ids].sample(n=50)
```

```python
# Seleccion final de la muestra de 50 registros
df_scoreo_oot_final_sample = df_scoreo_oot_final.sample(n=50, random_state=seed) 
df_scoreo_oot_final_sample
```

```python
# nos quedamos solo con las features que usamos en el entrenamiento del modelo para hacer el predict del modelo
X_oot = df_scoreo_oot_final_sample[final_features].copy()
```

## 3) Scoreo de la muestra final

```python
# Paso 1: Levantamos el modelo ya entrenado en la notebook anterior
import pickle
with open('model_catboost.pickle', 'rb') as f:
    model = pickle.load(f)
```

```python
# Paso 2: predict sobre la data nueva
pred = model.predict(X_oot)
```

```python
len(pred)
```

## 4) Armado del output final

```python
# Precio promedio en MELI
if foco_muestra in ['alta_presencia','sin_foco']:
    grouped = df_scoreo_oot_final.groupby(['match_marca_a','match_modelo_a','match_v1_a','car_year'], as_index=False)['price_meli_ok'].mean()
    grouped.rename(columns={'price_meli_ok':'precio_prom_meli'},inplace=True)
    df_scoreo_oot_final_sample = df_scoreo_oot_final_sample.merge(grouped, how='inner'
                                                                  , on=['match_marca_a','match_modelo_a','match_v1_a','car_year'])
elif foco_muestra == 'baja_presencia':  # si el foco es baja_presencia, la agrupación la hacemos de menos niveles
    grouped = df_scoreo_oot_final.groupby(['match_marca_a','match_modelo_a','car_year'], as_index=False)['price_meli_ok'].mean()
    grouped.rename(columns={'price_meli_ok':'precio_prom_meli'},inplace=True)
    df_scoreo_oot_final_sample = df_scoreo_oot_final_sample.merge(grouped, how='inner'
                                                                  , on=['match_marca_a','match_modelo_a','car_year'])
else:
    raise ValueError('Revisar el valor que le pasamos al parametro "foco_muestra", las opciones disponibles son: "alta_presencia", "baja_presencia" o "sin_foco"')
```
